```python
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/NLPCourse/NLPFinal/ProcessedPR.csv')
```

```python
import matplotlib.pyplot as plt

data['Press Release'] = data['Press Release'].fillna('')

data['Doc Length'] = data['Press Release'].apply(len)


data['Doc Length'] = data['Press Release'].apply(len)

# Visualization 1: Document Length Distribution
plt.figure(figsize=(10, 5))
plt.hist(data['Doc Length'], bins=30, color='skyblue')
plt.title('Distribution of Document Lengths')
plt.xlabel('Document Length')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Visualization 2: Party Membership Count
plt.figure(figsize=(7, 5))
party_counts = data['Party'].value_counts()
party_colors = ['red' if party == 'R' else 'cornflowerblue' for party in party_counts.index]
party_counts.plot(kind='bar', color=party_colors)
plt.title('Count of Members by Party')
plt.xlabel('Party')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
# Visualization: Box Plot of Document Lengths per Party
plt.figure(figsize=(10, 6))
data.boxplot(column='Doc Length', by='Party', grid=True, patch_artist=True, color=dict(boxes='Black', whiskers='Black', medians='Red', caps='Black'))
plt.title('Distribution of Document Lengths by Party')
plt.suptitle('')  # Remove the default title to clean up the plot
plt.xlabel('Party')
plt.ylabel('Document Length')
plt.show()
```

```python
# Calculate the average document length per party
average_doc_length_per_party = data.groupby('Party')['Doc Length'].mean()

# Visualization: Average Document Length per Party
plt.figure(figsize=(8, 5))
average_doc_length_per_party.plot(kind='bar', color=['cornflowerblue', 'red', 'green', 'purple', 'orange'])
plt.title('Average Document Length per Party')
plt.xlabel('Party')
plt.ylabel('Average Document Length')
plt.grid(True)
plt.show()
```

```python
data_caucus = pd.read_csv('/content/drive/MyDrive/NLPCourse/NLPFinal/ProcessedCaucus.csv')
```

```python
import matplotlib.pyplot as plt

# Handle missing values in 'Press Release'
data_caucus['Press Release'] = data_caucus['Press Release'].fillna('')

# Calculate the length of each press release
data_caucus['Doc Length'] = data_caucus['Press Release'].apply(len)

# Visualization 1: Document Length Distribution by Caucus
plt.figure(figsize=(10, 6))
data_caucus['Doc Length'].hist(by=data_caucus['Caucus'], bins=20, layout=(3, 3), color='skyblue', sharex=True, sharey=True)
plt.suptitle('Distribution of Document Lengths by Caucus')
plt.xlabel('Document Length')
plt.ylabel('Frequency')
plt.show()

# Visualization 2: Caucus Membership Count
plt.figure(figsize=(8, 5))
data_caucus['Caucus'].value_counts().plot(kind='bar', color='cornflowerblue')
plt.title('Count of Members by Caucus')
plt.xlabel('Caucus')
plt.ylabel('Count')
plt.grid(True)
plt.show()

# Visualization 3: Box Plot of Document Lengths by Caucus
plt.figure(figsize=(12, 7))
data_caucus.boxplot(column='Doc Length', by='Caucus', grid=True, patch_artist=True, color=dict(boxes='Black', whiskers='Black', medians='Red', caps='Black'))
plt.title('Distribution of Document Lengths by Caucus')
plt.suptitle('')
plt.xlabel('Caucus')
plt.ylabel('Document Length')
plt.show()
```
